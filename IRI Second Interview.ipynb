### Question 2: How to fill in the missing data for the group the transaction is associated with?

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.model_selection import StratifiedKFold
```

```python
X = merged_data.drop(columns=["Transaction_Associated_Group", "Transaction_Date", "Transaction_ID"]).to_numpy()
y = merged_data["Transaction_Associated_Group"].to_numpy()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
```

```python
classifier = RandomForestClassifier()
classifier.fit(X_train, y_train)
```

```python
folds = StratifiedKFold(n_splits=4)
```

```python
def get_score(model, X_train, X_test, y_train, y_test):
    model.fit(X_train, y_train)
    return model.score(X_test, y_test)
```

```python
scores_rf = []
random_forest = RandomForestClassifier(n_estimators=30)
for train_index, test_index in folds.split(X,y):
    X_train, X_test, y_train, y_test = X[train_index], X[test_index], \
                                       y[train_index], y[test_index]
    scores_rf.append(get_score(random_forest, X_train, X_test, y_train, y_test))
```
